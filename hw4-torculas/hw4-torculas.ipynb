### Importing Libraries<a class="anchor" id="libraries"></a>

```python
##### STANDARD LIBRARIES #####

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

##### PREPROCESSING #####
import os
import re
import email
import codecs

%matplotlib inline

##### PERFORMANCE EVALUATION #####
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import accuracy_score, recall_score, precision_score
```

## Preprocessing<a class="anchor" id="preprocess"></a>

### Extraction of Files and the Data<a class="anchor" id="extract"></a>

#### Creation of Dataframe<a class="anchor" id="dataframe"></a>

```python
df = pd.DataFrame(columns = ['folder', 'file', 'emailMessage', 'classification'])
```

```python
labelsPath = "trec06p-cs280/labels"
dfLabels = pd.read_csv(labelsPath, sep = " ", header = None)

# Dataframe for the labels, calssification and its respective filepath
dfLabels.columns = ["classification", "filePath"] 

# Spam and ham classification will be converted into 1 and 0, respectively.
dfLabels["classification"] = dfLabels["classification"].apply(lambda kv: 0 if kv == "ham" else 1)

# Remove "../data/" from filepath
dfLabels["filePath"] = dfLabels["filePath"].apply(lambda kv: kv.replace("../data/", ""))
dfLabels.head()
```

#### Creation of Data Directory<a class="anchor" id="directory"></a>

```python
# List of all the folders for data directory

foldersPath = "trec06p-cs280/data"
folders = sorted(os.listdir(foldersPath))
```

### Removal of Unecessary Contents Found in the Email<a class="anchor" id="remove"></a>

Remove words from the document which may not contribute to the information we want to extract. These includes dropping the alphanumeric characters and punctuation marks. Also, remove stop words, more popularly known as meaningless words, from the email body since those words are not useful in classification as well as reduce the dimensionality of the dictionary. stop_words.txt

#### Initialize the Characters<a class="anchor" id="initialize"></a>

```python
#Punctuation marks
punc = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~\\"
#Numbers
num = "0123456789"
#Escape characters
esc = re.compile(r'\\[a-z][a-z]?[0-9]+')
#HTML tags
tags = re.compile('<.*?>') 
#List of information found in the email that will be dropped
stop_words = []
#Get the stop words
with open('stop_words.txt', 'r') as file:
    stop_word = file.read().splitlines()
    stop_words = [word for word in stop_word]
```

#### Functions for Removing and Getting the Message From Parsed Email<a class="anchor" id="functions"></a>

```python
def removeWords(msg):
    msg = msg.lower()                                   # Convert the strings into lowercases
    msg = re.sub(tags, '', msg)                         # Remove HTML tags
    msg = msg.translate(str.maketrans('', '', punc))    # Remove punctuation marks
    msg = msg.translate(str.maketrans('', '', num))     # Remove numbers
    msg = re.sub(esc, '', repr(msg))                    # Remove escape characters
    msg = codecs.decode(msg, 'unicode_escape')          # Convert raw string to have unicode escape characters
    words = msg.split()                                 # Splits msg (string) into a list of words
    
    # Remove all stop words
    words = [word for word in words if word not in stop_words]
    msg = " ".join(words)
    msg = msg.strip("'")
    return msg

def getMessage(parsed):
    msg = ""
    if parsed.is_multipart():           # Multipart email
        for part in parsed.walk():      # Iterate email
            #If the content type of the part is text/plain,
            if part.get_content_type() == 'text/plain':
                #Get the message
                msg = part.get_payload()
                break
    else:
        msg = parsed.get_payload()
    return msg
```

#### Accessing the File and Storing into the Dataframe<a class="anchor" id="access"></a>

```python
for folder in folders:
    files = sorted(os.listdir(f"{foldersPath}/{folder}"))      # Get the files inside the folder
    for file in files:
        with open(f"{foldersPath}/{folder}/{file}", "r", encoding = "ISO-8859-1") as e_mail:
            read_email_file = e_mail.read()
            parsed = email.message_from_string(read_email_file)
            msg = getMessage(parsed)
            msg = removeWords(msg)
            
            #Get the classification of the email based on the Labels dataframe
            labels_classification = dfLabels[dfLabels['filePath'] == f"{folder}/{file}"]['classification'].values[0]
            
            #Concatenate the data to the main dataframe
            df = pd.concat([df, pd.DataFrame([[folder, file, msg, labels_classification]], columns = ["folder", "file", "emailMessage", "classification"])], ignore_index = True)
df
```

```python
if not os.path.exists('preprocessedFiles'):
    os.makedirs('preprocessedFiles')
df.to_csv("preprocessedFiles/preprocessedEmails.csv", index = False)
#Reset the main dataframe for memory overload to not occur
df.drop(df.index, inplace = True)
df = pd.DataFrame(columns = ["folder", "file", "emailMessage", "classification"])
```

### Importing the Preprocessed Data<a class="anchor" id="import"></a>

#### Splitting Dataset<a class="anchor" id="split"></a>

```python
df2 = pd.read_csv("./preprocessedFiles/preprocessedEmails.csv")

traindf = df2[df2['folder'] < 71]     # Folders 0-70: Train Set
testdf = df2[df2['folder'] >= 71]     # Folders 71-127: Test Set

trainingHam = traindf[traindf['classification'] == 0]
trainingSpam = traindf[traindf['classification'] == 1]
```

#### Extracting List of Unique Words<a class="anchor" id="extract"></a>

Extract a list of unique words from the training set along with its summed number of occurrences from the spam and ham set. To limit the cardinality of the dictionary, we can extract only the 10000 most common words (common means that these words have the highest frequencies/occurences in the dataset)

```python
wordsDict = {}

for index, row in traindf.iterrows():
    for word in str(row['emailMessage']).split():
        if word in wordsDict:
            wordsDict[word] += 1
        else:
            wordsDict[word] = 1

#Sort the dictionary by value in descending order
sortedWordsDict = sorted(wordsDict.items(), key = lambda kv: kv[1], reverse = True)

#Extracting the 10,000 most common words
mostCommonWords = dict(sortedWordsDict[:10000])
mostCommonWords
```

## Creating the Feature Matrices<a class="anchor" id="featurematrices"></a>

Feature matrices for both the spam training set and ham training set with a dimensionality of 10000 (this was the specified number of different words in the dictionary). For each word in the dictionary, traverse through each file and check its occurrence. 1 denotes the existence of a word in the email and 0 otherwise. Thus, a matrix whose rows denote the number of files of training/testing set and columns denoting 10000 words will be generated. Every email in the training as a feature vector.

Consider that the higher the cardinality of the vocabulary, the higher the dimensionality of the matrix. The matrix grows with respect to the number of emails considered. Hence, slowing the program.

### Feature Matrix of Spam Training Set<a class="anchor" id="fmspam"></a>

```python
featureMatrixSpam = np.zeros((len(trainingSpam), 10000))
listMostCommonWords = list(mostCommonWords.keys())

for index in range(len(trainingSpam)):
    for word in str(trainingSpam.iloc[index]['emailMessage']).split():
        if word in mostCommonWords:
            featureMatrixSpam[index][listMostCommonWords.index(word)] = 1
            
featureMatrixSpam
```

### Feature Matrix of Ham Training Set<a class="anchor" id="fmham"></a>

```python
featureMatrixHam = np.zeros((len(trainingHam), 10000))
listMostCommonWords = list(mostCommonWords.keys())

for index in range(len(trainingHam)):
    for word in str(trainingHam.iloc[index]['emailMessage']).split():
        if word in mostCommonWords:
            featureMatrixHam[index][listMostCommonWords.index(word)] = 1
            
featureMatrixHam
```

## Computing the Priors<a class="anchor" id="computepriors"></a>

The prior probabilities for spam and ham are computed by the following formula:

$ P(c=ham)= {N_{ham} \over N_{doc}}$

$ P(c=spam)= {N_{spam} \over N_{doc}}$

Where $N_{ham}$ is the number of spam emails in the training set, $N_{spam}$ is the number of spam emails in the training set, and $N_{doc}$ is the total number of emails

```python
hamEmails = len(trainingHam)       # Number of ham emails in the training set
spamEmails = len(trainingSpam)     # Number of spam emails in the training set
docEmails = len(traindf)           # Total number of emails

# Formulas
p_ham = hamEmails/docEmails
p_spam = spamEmails/docEmails

print(f"P(c = ham) = {p_ham}")
print(f"P(c = spam) = {p_spam}")
```

## Computing the Likelihood of Each Word with Laplace Smoothing<a class="laplace" id="application"></a>

A vector containing the number of occurrences of each word in the dataset will be created. One vector will contain the number of occurrences of each word in the ham vocabulary and another vector for the spam vocabulary.

```python
def laplaceSmoothing(featureMatrixSpam, featureMatrixHam):
    # Initializing the probability of each word 
    probSpam = np.zeros(len(mostCommonWords))
    probHam = np.zeros(len(mostCommonWords))
    
    #Word count of spam and ham
    wordCountSpam = np.sum(featureMatrixSpam, axis = 0)
    wordCountHam = np.sum(featureMatrixHam, axis = 0)
    
    #Total number of words in spam and ham
    totalWordsSpam = np.sum(wordCountSpam)
    totalWordsHam = np.sum(wordCountHam)
    
    param = 1
    num_of_classes = 2

    # Computing the likelihood of each word with laplace smoothing
    for i in range(len(mostCommonWords)):
        probSpam[i] = (wordCountSpam[i] + param)/(totalWordsSpam + param*num_of_classes)
        probHam[i] = (wordCountHam[i] + param)/(totalWordsHam + param*num_of_classes)
    return probSpam, probHam

likelihoodSpam, likelihoodHam = laplaceSmoothing(featureMatrixSpam, featureMatrixHam)

print(f"Likelihood Spam with laplace smoothing: {likelihoodSpam}")
print(f"Likelihood Ham with laplace smoothing: {likelihoodHam}")
```

## Classifying the Emails: Computing the Log Probabilities<a class="anchor" id="classify"></a>

Based on the probabilities of words given the label computed, determine the probability of ham and the probability of spam given the document. Whichever has a higher value will be the classification of the email. Baye’s rule was applied to the computed likelihoods.

```python
def classifyingEmails(email, likelihoodHam, likelihoodSpam, p_ham, p_spam):
    logProbHam = 0
    logProbSpam = 0

    words = str(email).split()         # Splitting the email into words
    for word in words:
        if word in mostCommonWords:
            logProbHam += np.log(likelihoodHam[listMostCommonWords.index(word)])
            logProbSpam += np.log(likelihoodSpam[listMostCommonWords.index(word)])
            
    logProbHam += np.log(p_ham)
    logProbSpam += np.log(p_spam)
    
    if logProbHam > logProbSpam:
        return 0
    else:
        return 1
    
#Sorting train emails
traindf['pred'] = traindf['emailMessage'].apply(lambda kv: classifyingEmails(kv, likelihoodHam, likelihoodSpam, p_ham, p_spam))
```

```python
count = 0
for index, row in traindf.iterrows():
    if float(row['classification']) == float(row['pred']):
        count += 1
print(" Emails: {} \n Emails Classified Correctly: {} \n Percentage of Emails Correctly Classified: {}%.".format(len(traindf), count, count/len(traindf)*100))
```

### Testing the Classifier<a class="anchor" id="test"></a>

```python
testdf['pred'] = testdf['emailMessage'].apply(lambda kv: classifyingEmails(kv, likelihoodHam, likelihoodSpam, p_ham, p_spam))

testdf
```

```python
count = 0
for index, row in testdf.iterrows():
    if float(row['classification']) == float(row['pred']):
        count += 1
print(" Emails: {} \n Emails Classified Correctly: {} \n Percentage of Emails Correctly Classified: {}%.".format(len(testdf), count, count/len(testdf)*100))
```

## Performance Evaluation<a class="anchor" id="evaluate"></a>

Confusion Matrix

#### False Positive Rate, False Negative Rate, True Positive Rate, True Negative Rate<a class="anchor" id="all"></a>

```python
actual = np.array(testdf['classification'])
predicted = np.array(testdf['pred'])

confusion_matrix = metrics.confusion_matrix(actual, predicted, labels = [0, 1])

sns.heatmap(confusion_matrix, annot = True, fmt = 'd')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

print("False Positive Rate (FP) - {}".format(confusion_matrix[0][1])) 
print("False Negative Rate (FN) - {}".format(confusion_matrix[1][0])) 
print("True Positive Rate (TP) - {}".format(confusion_matrix[1][1])) 
print("True Negative Rate (TN) - {}".format(confusion_matrix[0][0])) 
```

#### Accuracy, Recall, Precision<a class="anchor" id="accuracy"></a>

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score

accuracy = accuracy_score(actual, predicted)
recall = recall_score(actual, predicted)
precision = precision_score(actual, predicted)

print("Accuracy - {}".format(accuracy))
print("Recall - {}".format(recall))
print("Precision - {}".format(precision))
```
